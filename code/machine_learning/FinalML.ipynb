### Function to split and visualize the data
To start, let's use all previous course years to try and predict 2015 (our last year for which we have target values). Since we will eventually want to use model to predict future scores, this seems like the most relevant starting point

```python
def visual_model_selection(df, test_year, model):
    
    X_train = df[df.m1_fall != test_year].drop(['m1_fall', 'target'], 1)
    X_test  = df[df.m1_fall == test_year].drop(['m1_fall', 'target'], 1)
    y_train = df[df.m1_fall != test_year][['target']]
    y_test  = df[df.m1_fall == test_year][['target']]
    
    visualizer = ClassificationReport(model, classes=['Not At Risk', 'At Risk'])
    visualizer.fit(X_train, y_train)  
    visualizer.score(X_test, y_test)
    visualizer.poof()
```

```python
def LoopYears(df, mods = models):

    print('Generating Classification Reports')
    for year in years:
        num_targets = df[(df.target == 1) & 
            (df.m1_fall == year)].count()
        num_targets = num_targets[1]
    
        num_students = df[df.m1_fall == year].count()
        num_students = num_students[1]
    
        print('\n----------------------------------------------')
        print('Year to be tested : [' +  str(year) + ']')
        print('Total Targets students for this year: [' + str(num_targets) + ']')
        print('Total Students for this year: [' + str(num_students) + ']')
        print('----------------------------------------------\n')
    
        for mod in mods.keys():
            visual_model_selection(df, year, mods[mod]) 
    
```

My last attempt at getting something better will be to take the zscore of the courses by year. I thought this might be overkill before, but maybe it will help. 

```python
def yearly_zscore(df, columns, keep, year_var = 'year'):
    for c in columns:
        df[c + '_mean'] = df.groupby(year_var)[c].transform('mean')
        df[c + '_std']  = df.groupby(year_var)[c].transform('std')
        df[c]    = (df[c] - df[c + '_mean'])/(df[c + '_std'])
        if keep == False:
            df = df.drop([c + '_mean', c + '_std'], 1)
    return df
```

### Let's also try and tune class weights. I just want to do this based  on the AT RISK F1 score. 

```python
def f1_score(df, test_year, model):
    
    X_train = df[df.m1_fall != test_year].drop(['m1_fall', 'target'], 1)
    X_test  = df[df.m1_fall == test_year].drop(['m1_fall', 'target'], 1)
    y_train = df[df.m1_fall != test_year][['target']]
    y_test  = df[df.m1_fall == test_year][['target']]
    
    fitted_model = model.fit(X_train, y_train)
    fitted_model.score(X_test, y_test)
    expected   = y_test
    predicted  = fitted_model.predict(X_test)

    classificationReport = classification_report(expected, predicted, target_names=["Not At Risk","At Risk"])
    scores = [recall_score(expected, predicted), precision_score(expected, predicted)]
    return scores
```

```python
def max_weights(df = Final_data, start = 1, end = 5, interval = .01, test_year = 2015 , model = LogisticRegression()):

    wgt = start 
    if interval < .01:
        interval = .01
    wgt_scores = {}
        
    while wgt < end:
        weight_dic = {0:1, 1:wgt}
        scores = f1_score(Final_data, test_year, LogisticRegression(class_weight = weight_dic))
        F1 = 2*(scores[0]*scores[1])/(scores[0] + scores[1])
        wgt_scores[(wgt)] = F1
        wgt = wgt + interval
        
    max_score = 0
    max_dict = ''

    for key in wgt_scores.keys():
        if wgt_scores[key] > max_score:
            max_score = wgt_scores[key]
            max_dict = key
    return [max_score, max_dict]
        
```

```python
yearly_scores = []
for yr in years:
    score = max_weights(test_year = yr, model = LogisticRegression(C = 2.78))
    print(score)
    yearly_scores.append(score[1])
```
