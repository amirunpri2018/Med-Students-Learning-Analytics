### Function to split and visualize the data
To start, let's use all previous course years to try and predict 2015 (our last year for which we have target values). Since we will eventually want to use model to predict future scores, this seems like the most relevant starting point

```python
def visual_model_selection(df, test_year, model):
    
    X_train = df[df.m1_fall != test_year].drop(['m1_fall', 'target'], 1)
    X_test  = df[df.m1_fall == test_year].drop(['m1_fall', 'target'], 1)
    y_train = df[df.m1_fall != test_year][['target']]
    y_test  = df[df.m1_fall == test_year][['target']]
    
    visualizer = ClassificationReport(model, classes=['Not At Risk', 'At Risk'])
    visualizer.fit(X_train, y_train)  
    visualizer.score(X_test, y_test)
    visualizer.poof()
```

## Part Five: Let's look at one clean summary table
Those tables above are too hard to look at. Let's just focus on F1 score and create a heatmap

```python
def model_f1(df, test_year, model):
    
    X_train = df[df.m1_fall != test_year].drop(['m1_fall', 'target'], 1)
    X_test  = df[df.m1_fall == test_year].drop(['m1_fall', 'target'], 1)
    y_train = df[df.m1_fall != test_year][['target']]
    y_test  = df[df.m1_fall == test_year][['target']]
    
    model.fit(X_train, y_train)
    expected = y_test
    predicted = model.predict(X_test)
    return(list(f1_score(expected, predicted, average = None))[1])
```

```python
def get_f1_df(df, models, years):
    all_scores = {}
    for year in years:
        mod_scores = {}
        for mod in models.keys():
            mod_scores[mod] = round(model_f1(df, year, models[mod]), 3)    
        all_scores[year] = mod_scores
    score_table = pd.DataFrame(all_scores)
    score_table['Average'] = score_table.mean(axis = 1)
    return score_table
```

```python
def f1_heatmap(df, models, years, save = False):
    data = get_f1_df(df, models, years)
    sizes ={'axes.labelsize': 32, 'font.size': 32, 'legend.fontsize': 32.0, 'axes.titlesize': 32}
    a4_dims = (10, 8)
    fig, ax = plt.subplots(figsize=a4_dims)
    sns.set(font_scale=1.3) 
    ax = sns.heatmap(data, cmap = "Blues",  annot = True, linewidths = 0.5)#, vmin =0, vmax = 1)
    ax.set_yticklabels(ax.get_yticklabels(), rotation =45)
    ax.set_title('At-Risk F1 Score Heat Map')
    if save == True:
        fig.savefig('./visuals/heatmap') 
```

```python
f1_heatmap(Final_data, models, years)
```

My last attempt at getting something better will be to take the zscore of the courses by year. I thought this might be overkill before, but maybe it will help. 

```python
def yearly_zscore(df, columns, keep, year_var = 'year'):
    for c in columns:
        df[c + '_mean'] = df.groupby(year_var)[c].transform('mean')
        df[c + '_std']  = df.groupby(year_var)[c].transform('std')
        df[c]    = (df[c] - df[c + '_mean'])/(df[c + '_std'])
        if keep == False:
            df = df.drop([c + '_mean', c + '_std'], 1)
    return df
```

## Part Six: Hyperparameter Tuning

Let's do a grid search. We can sort of see that LogisticRegression is our best model. We've done some hyperparameter tuning in the other notebooks, but nothing formal like a grid search. 

```python
X = Final_data[ courses + indicators + continuous + discontinuous]
y = Final_data['target']
```

### Class Weight
The first thing we should gridsearch is class weight because we know it's the most important

```python
def get_f1_score(df, test_year, model):
    
    X_train = df[df.m1_fall != test_year].drop(['m1_fall', 'target'], 1)
    X_test  = df[df.m1_fall == test_year].drop(['m1_fall', 'target'], 1)
    y_train = df[df.m1_fall != test_year][['target']]
    y_test  = df[df.m1_fall == test_year][['target']]
    
    model.fit(X_train, y_train)
    expected   = y_test
    predicted  = model.predict(X_test)
    scores = list(f1_score(expected, predicted, average = None))[1]
    return scores
```

```python
def max_weights(df = Final_data, start = 1, end = 5, interval = .01, test_year = 2015 , model = LogisticRegression()):

    wgt = start 
    if interval < .01:
        interval = .01
    wgt_scores = {}
        
    while wgt < end:
        weight_dic = {0:1, 1:wgt}
        F1 = get_f1_score(Final_data, test_year, model.set_params(class_weight = weight_dic))
        wgt_scores[(wgt)] = F1
        wgt = wgt + interval
        
    max_score = 0
    max_dict = ''

    for key in wgt_scores.keys():
        if wgt_scores[key] > max_score:
            max_score = wgt_scores[key]
            max_dict = key
    return [max_score, max_dict]
```

```python
def find_best_weights(model, years = years):
    yearly_scores = []
    for yr in years:
        print(yr)
        score = max_weights(test_year = yr, model = model)
        yearly_scores.append(score[1])
    print(sum(yearly_scores)/float(len(yearly_scores))) 
```

#### Best Weight: Logistic Regression

```python
find_best_weights(LogisticRegression())
```

#### Best Weight: Linear SVC

```python
find_best_weights(LinearSVC())
```

#### Best Weight: SVC

```python
find_best_weights(SVC())
```

### GridSearch Other Parameters based on weight

```python
def GridSearch(model, parameters):
    hyperparameters = parameters
    clf = GridSearchCV(model, hyperparameters, cv=5, verbose=0)
    best_model = clf.fit(X, y)
    #print('Best Penalty:', best_model.best_estimator_.get_params()['penalty'])
    #print('Best C:', best_model.best_estimator_.get_params()['C'])
    return best_model.best_estimator_.get_params()
    
```

#### GridSearch: LogisticRegression()

```python
Cs = np.linspace(0.1,5,20)
LogReg_params = dict(C = Cs)
GridSearch(LogisticRegression(class_weight = {0:1, 1:2.76}), parameters = LogReg_params)
```

#### GridSearch Linear SVC

```python
Cs = np.linspace(1, 5,20)
LSVC_params = dict(C = Cs)
GridSearch(LinearSVC(class_weight = {0:1, 1:2.487}), LSVC_params)
```

#### GridSearch SVC

```python
Cs = np.linspace(0.1,4,20)
SVC_params = dict(C = Cs)
GridSearch(SVC(class_weight = {0:1, 1:2.45}), SVC_params)
```
