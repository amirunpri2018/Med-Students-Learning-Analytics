# Initial Data Query and Visualization

### Import Packages and Set Directories

```python
import pandas   as pd
import numpy    as np
import sqlite3  as sql

code_dir = "/Users/Zach/data_science/Med-Students-Learning-Analytics/code"
csv_dir = "/Users/Zach/data_science/med_school_data/csvs/"
sql_dir = "/Users/Zach/data_science/med_school_data/sql/"
```

### Connect the cursor

```python
con = sql.connect(sql_dir + "medschool.sqlite")
cur = con.cursor()
```

### Write the Query

```python
query1 = '''
					select person_uid_anon, count(person_uid_anon) as count
					from grades 
					group by person_uid_anon
					order by count desc
					;
				'''				
```

### Execute the Queries

```python
number_of_students = pd.read_sql(query1, con)
number_of_students["person_uid_anon"].nunique
number_of_students
```

### Plot 1: Student Histogram
#### Description: A histogram of how often each student ID shows up in the grades table

```python

number_of_students.hist(column = "count")
```
